```python
df = pd.read_csv("Mobile_Reviews_Sentiment.csv")
```

```python
df['sentiment'] = df['sentiment'].str.strip().str.lower()

```

```python
def get_sentiment(rating):
    if rating >= 4:
        return 'positive'
    elif rating == 3:
        return 'neutral'
    else:
        return 'negative'

df['sentiment'] = df['rating'].apply(get_sentiment)
```

```python
print(df[['rating', 'sentiment']].head(10))
```

```python
df['language'] = df['language'].str.strip().str.lower()
```

```python
language_map = {
    'english': 'en',
    'en': 'en',
    'eng': 'en',
    'spanish': 'es',
    'es': 'es',
    'esp': 'es',
    'french': 'fr',
    'fr': 'fr',
    'german': 'de',
    'de': 'de',
    'italian': 'it',
    'it': 'it',
    'hindi': 'hi',
    'hi': 'hi',
    'chinese': 'zh',
    'zh': 'zh',
    'mandarin': 'zh'
}
```

```python
df['language'] = df['language'].replace(language_map)
```

```python
valid_iso_codes = ['en', 'es', 'fr', 'de', 'it', 'hi', 'zh']
df = df[df['language'].isin(valid_iso_codes)]
```

```python
df.to_csv("Mobile_Reviews_Sentiment_clean.csv", index=False)
```

```python
print("Language column standardized to valid ISO codes!")
```

```python
df.head()
```

```python
df['year'] = df['review_date'].dt.year
df['month'] = df['review_date'].dt.month
df['day'] = df['review_date'].dt.day 
print(['df'])
```

```python
df.head
```

```python
#4a, NUMERICAL CONSISTENCY
def clean_price(x):
    if isinstance(x, str):
        x = x.replace(",", "").replace("₹", "").replace("R$", "").replace("د.إ", "").strip()
    try:
        return float(x)
    except:
        return np.nan

df['price_usd'] = df['price_usd'].apply(clean_price)
df['price_local'] = df['price_local'].apply(clean_price)
df['exchange_rate_to_usd'] = pd.to_numeric(df['exchange_rate_to_usd'], errors='coerce')

```

```python
tolerance = 0.01
price_check = np.isclose(df['price_local'] / df['exchange_rate_to_usd'], df['price_usd'], atol=tolerance)
misaligned_prices = df[~price_check]
print(df[~np.isclose(df['price_local'] / df['exchange_rate_to_usd'], df['price_usd'], atol=0.01)])
```

```python
#4b, NUMERICAL CONSISTENCY
rating_columns = [
    'battery_life_rating',
    'camera_rating',
    'performance_rating',
    'design_rating',
    'display_rating'
]

invalid_ratings = df[
    (df[rating_columns] < 1).any(axis=1) |
    (df[rating_columns] > 5).any(axis=1)
]
print(invalid_ratings)
```

```python
# Validate review_length and word_count against review_text
mismatch_review_length = df[df['review_length'] != df['review_text'].str.len()]
mismatch_word_count = df[df['word_count'] != df['review_text'].str.split().str.len()]
```

```python
# 4d, HELPFUL VOTES
df["helpful_votes"] = df["helpful_votes"].astype(int).clip(lower=0)
```

```python
# Standardize brand and model names
df['brand'] = df['brand'].str.strip().str.title()
df['model'] = df['model'].str.strip().str.title()
```

```python
#Verify currency
# Check valid currency values
valid_currencies = {'USD', 'INR', 'BRL', 'AED'}
invalid_currency = df.loc[~df['currency'].isin(valid_currencies), 'currency']
print(invalid_currency)
```

```python
#6a, NEW FEATURES
df['verified_purchase_binary'] = df['verified_purchase'].astype(int)
print(df['verified_purchase_binary'])
```

```python
#6b, review sentiment score
def sentiment_score(sentiment):
    if sentiment =='Positive':
        return 1
    elif sentiment == 'Neutral':
        return 0
    else:
        return 1

df['review_sentiment'] = df['sentiment'].apply(sentiment_score)
print(df['review_sentiment'])
```

```python
#6c, Average rating
rating_columns = ['battery_life_rating', 'camera_rating', 'performance_rating', 'design_rating','display_rating']  # example columns
df['average_rating'] = df[rating_columns].mean(axis=1)
print(df[['average_rating']].head())
```

```python
#6d, NEW FEATURES
def length_category(count):
    if count < 10:
        return 'short'
    elif count == 10:
        return 'medium'
    else:
        return 'long'

df['review_length'] = df['word_count'].apply(length_category)
print(df['review_length'])
```
